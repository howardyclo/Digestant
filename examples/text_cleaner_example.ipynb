## Get text from data

```python
texts = df['text']
```

## Uncleaned text

```python
for i, text in enumerate(texts):
    print('Text {}:'.format(i))
    print(text)
    print('-'*116)
    
    if i == 4: break
```

## Let's clean it

```python
text_cleaner = TextCleaner(filter_sentiment_words=True) # default value of `filter_sentiment_words` is false
```

```python
cleaned_texts = text_cleaner.clean(texts)
```
